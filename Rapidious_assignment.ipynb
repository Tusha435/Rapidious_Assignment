```python
# Percentage of missing values in each column
missing_percentage = (data.isnull().sum() / len(data)) * 100
print(missing_percentage)
```

```python
# Threshold for dropping columns with excessive missing data 
threshold = 40
columns_to_drop = missing_percentage[missing_percentage > threshold].index
columns_to_drop
```

```python
# Drop columns with more than 40% missing data
data_cleaned = data.drop(columns=columns_to_drop)

# Drop duplicate rows
data_cleaned = data_cleaned.drop_duplicates()
```

```python
# Recheck missing values and shape after cleaning
remaining_missing = data_cleaned.isnull().sum()
cleaned_shape = data_cleaned.shape
```

```python
# For numerical columns, using mean imputation for simplicity
data_imputed = data_cleaned.copy()
numerical_columns = ['calories', 'protein', 'fat']

for col in numerical_columns:
    data_imputed[col].fillna(data_imputed[col].mean(), inplace=True)
```

```python
print(data_imputed)
```

```python
# Function to identify outliers based on the IQR method
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

# Identifying outliers in calories, protein, and fat columns
calories_outliers = detect_outliers(data_imputed, 'calories')
protein_outliers = detect_outliers(data_imputed, 'protein')
fat_outliers = detect_outliers(data_imputed, 'fat')

# Count of outliers in each column
outliers_count = {
    'calories': len(calories_outliers),
    'protein': len(protein_outliers),
    'fat': len(fat_outliers)
}
print("Outlier Count")
outliers_count
```
